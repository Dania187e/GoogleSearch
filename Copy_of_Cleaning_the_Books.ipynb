```python
import torch
from transformers import pipeline

pipe = pipeline("text-generation", model="HuggingFaceH4/zephyr-7b-alpha", torch_dtype=torch.bfloat16, device_map="auto")

```

```python
from tqdm import tqdm
tokenizer = pipe.tokenizer

def split_into_paragraphs(text, max_length=512):
    paragraphs = []
    current_paragraph = ""

    tokens = tokenizer.tokenize(tokenizer.decode(tokenizer.encode(text)))

    for token in tqdm(tokens):
        if len(tokenizer.encode(current_paragraph + token)) <= max_length:
            current_paragraph += token + " "
        else:
            paragraphs.append(current_paragraph.strip())
            current_paragraph = token + " "

    if current_paragraph:
        paragraphs.append(current_paragraph.strip())

    return paragraphs

book_text =data
paragraphs = split_into_paragraphs(book_text)

```

```python
from tqdm import tqdm
results_df = pd.read_csv('/content/drive/MyDrive/center_book/results/Dania.csv')

# Use len(paragraphs) as the total for tqdm
for i in tqdm(range(len(results_df), len(paragraphs))):

  messages = [
    {
        "role": "system",
        "content": "Given the following text, clean and extract sentences suitable for fine-tuning a medical model,the output in form of one sentence to give to a model for fine tune,if the text dosent contain any relvant or important medical information return _",

    },
    {"role": "user", "content": f"{paragraphs[i]}"},
  ]
  x = pipe.tokenizer.apply_chat_template(messages, tokenize=False, add_generation_prompt=True)
  output = pipe(x, max_new_tokens=512, do_sample=True, temperature=0.7, top_k=50, top_p=0.95)
  index = output[0]['generated_text'].find('<|assistant|>')
  output =output[0]['generated_text'][index+15:-1]

  # Append the output to the DataFrame
  results_df.loc[len(results_df.index)] = [output]  # Append as a new row
  results_df.to_csv('/content/drive/MyDrive/center_book/results/Dania.csv',index=False)
```
